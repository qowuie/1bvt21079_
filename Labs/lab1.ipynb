### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для

```python
import random
import time
```

```python
user_m = int(input())
user_n = int(input())
user_min_limit = int(input())
user_max_limit = int(input())
matrix = []
for m in range(0, user_m):
  matrix.append([])
for m in range(0, user_m):
  for n in range(0, user_n):
    matrix[m].append(random.randrange(user_min_limit, user_max_limit))

for line in matrix:
  print(line)
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()

def selection_sort(line):
  index = 0
  while index < len(line) - 1:
    min = index #индекс минимального элемента
    j = index + 1
    while j < len(line):
      if line[min] > line[j]:
        min = j
      j += 1
    line[index], line[min] = line[min], line[index]
    index += 1
  return line

mat = copy.deepcopy(matrix)
for line in mat:
  print(selection_sort(line))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
start_time = time.time()
def insertion_sort(line):
  for n in range(1, len(line)):
    i = line[n]
    j = n - 1
    while j >= 0 and i < line[j]:
      line[j + 1] = line[j]
      j -= 1
    line[j + 1] = i
  return line


mat2 = copy.deepcopy(matrix)
for line in mat2:
  print(insertion_sort(line))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()
def puzir(line):
  for n in range(0, len(line)):
    for j in range(len(line) - n - 1):
      if line[j] > line[j + 1]:
        line[j], line[j + 1] = line[j + 1], line[j]
  return line


mat3 = copy.deepcopy(matrix)
for line in mat:
  print(puzir(line))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
start_time = time.time()

def shell_sort(line):
  step = len(line) // 2
  while step > 0:
    for i in range(step, len(line)):
      elem = line[i]
      j = i
      while j >= step and line[j - step] > elem:
        line[j] = line[j - step]
        j -= step
      line[j] = elem
    step //= 2
  return line


mat4 = copy.deepcopy(matrix)
for line in mat:
  print(shell_sort(line))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
start_time = time.time()
def quick_sort(line):
  if len(line) <= 1:
    return line
  e = random.choice(line) # опорный элемент
  left = [i for i in line if i < e]
  center = [i for i in line if i == e]
  right = [i for i in line if i > e]
  return quick_sort(left) + center + quick_sort(right)

mat5 = copy.deepcopy(matrix)
for line in mat:
  print(quick_sort(line))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка.
start_time = time.time()
def heapify(line, heap_size, root):  
    l = root
    left = (2 * root) + 1
    right = (2 * root) + 2
    if left < heap_size and line[left] > line[l]:
        l = left
    if right < heap_size and line[right] > line[l]:
        l = right
    if l != root:
        line[root], line[l] = line[l], line[root]
        heapify(line, heap_size, l)

def heap(line):  
    size = len(line)
    for i in range(size, -1, -1):
        heapify(line, size, i)
    for i in range(size - 1, 0, -1):
        line[i], line[0] = line[0], line[i]
        heapify(line, i, 0)
    return line


mat6 = copy.deepcopy(matrix)
for line in mat:
  print(heap(line))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

### Вывод


```python
import matplotlib.pyplot as plt

def generate_matr(user_m=None, user_n=None, user_min_limit=None, user_max_limit=None):
  if user_m is None and user_n is None and user_min_limit is None and user_max_limit is None:
    user_m = int(input())
    user_n = int(input())
    user_min_limit = float(input())
    user_max_limit = float(input())
  matrix = []
  for m in range(0, user_m):
    matrix.append([])
  for m in range(0, user_m):
    for n in range(0, user_n):
      matrix[m].append(random.randrange(user_min_limit, user_max_limit))
  return matrix

def test(sort_method):
  elements = list()
  times = list()
  for i in range(1, 101):

      matr7 = generate_matr(user_m=i, user_n=i, user_min_limit=1, user_max_limit=20)
      start = time.time()
      for line in matr7:
        sort_method(line)
      end = time.time()

      elements.append(len(matr7[0]) * len(matr7))
      times.append(end-start)

  plt.xlabel('Matrix Size')
  plt.ylabel('Time')
  plt.plot(elements, times, label = sort_method.__name__)
  plt.grid()
  plt.legend()
  plt.show()

for func in (inbuilt, selection_sort, insertion_sort, puzir, 
             quick_sort, shell_sort, heap):
  test(func)
```
